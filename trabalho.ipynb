```python
#Importando Bibliotecas uteis
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importando Arquivo CSV
dados_df = pd.read_csv('bolsas2.csv', sep=';')
```

```python
#Removendo Celulas Vazias (Tratando os dados)
dados_df = dados_df.dropna()

#Exibindo os dados Tratados
display(dados_df)
```

```python
indice_maior = dados_df['Valor'].idxmax()
valor_maior = dados_df.loc[indice_maior, 'Valor']
nome_maior = dados_df.loc[indice_maior, 'Titulo']
print(nome_maior)
print(valor_maior)
```

```python
#Filtrar os valores diferentes de 0
valores_filtrados = dados_df[dados_df['Valor'] != 0]['Valor']

#Calcular a soma e a quantidade de valores filtrados
soma = valores_filtrados.sum()
quantidade = len(valores_filtrados)

#Calcular a média
media = soma / quantidade

#Exibir a média arredondada
print(round(media, 0))
```

```python
# Exibir os valores únicos na coluna 'AgFinanciadora'
valores_unicos = dados_df['AgFinanciadora'].unique()

# Exibir os valores únicos
print(valores_unicos)
```

```python
# Exibir os valores únicos na coluna 'Valor'
valores_unicos = dados_df['Valor'].unique()

# Exibir os valores únicos
print(valores_unicos)
```

```python
#Contagem dos registros para cada categoria de 'AgFinanciadora'
quantidade_volutaria = dados_df[dados_df['AgFinanciadora'] == 'Voluntário'].shape[0]
quantidade_ifes = dados_df[dados_df['AgFinanciadora'] == 'Ifes'].shape[0]
quantidade_CNPq = dados_df[dados_df['AgFinanciadora'] == 'CNPq'].shape[0]
quantidade_Fapes = dados_df[dados_df['AgFinanciadora'] == 'Fapes'].shape[0]

#Definir os dados para o gráfico
categorias = ['Voluntário', 'Ifes', 'CNPq', 'Fapes']
quantidades = [quantidade_volutaria, quantidade_ifes, quantidade_CNPq, quantidade_Fapes]
```

```python
#Imprimindo as quantidades
print("Quantidades de bolsas por AgFinanciadora:")
print(f"Voluntário: {quantidade_volutaria}")
print(f"Ifes: {quantidade_ifes}")
print(f"CNPq: {quantidade_CNPq}")
print(f"Fapes: {quantidade_Fapes}")
```

```python
#Criar o gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(categorias, quantidades, color='skyblue')

#Adicionar título e rótulos aos eixos
plt.title('Quantidade de Bolsas por AgFinanciadora', fontsize=14)
plt.xlabel('AgFinanciadora', fontsize=12)
plt.ylabel('Quantidade de Bolsas', fontsize=12)

#Adicionar os valores em cima de cada barra
for bar in bars:
    yval = bar.get_height()  # Pega a altura da barra (o valor)
    plt.text(bar.get_x() + bar.get_width()/2, yval, str(yval), ha='center', va='bottom', fontsize=10, color='black')

#Exibir o gráfico
plt.grid(True, axis='y', linestyle='--', alpha=0.7)  # Adiciona grade horizontal
plt.show()
```

```python
#Contar a quantidade de bolsas (registros) para cada campus
quantidade_bolsas_campus = dados_df.groupby('Campus').size()

#Encontrar o campus com maior quantidade de bolsas
campus_mais_bolsas = quantidade_bolsas_campus.idxmax()
quantidade_campus_mais_bolsas = quantidade_bolsas_campus.max()

#Imprimir o campus com maior quantidade de bolsas e seu respectivo número
print(f"O campus que mais recebeu bolsas foi: {campus_mais_bolsas}")
print(f"Quantidade de bolsas no campus {campus_mais_bolsas}: {quantidade_campus_mais_bolsas}")

#Criar o gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(quantidade_bolsas_campus.index, quantidade_bolsas_campus.values, color='lightcoral')

#Adicionar título e rótulos aos eixos
plt.title('Quantidade de Bolsas por Campus', fontsize=14)
plt.xlabel('Campus', fontsize=12)
plt.ylabel('Quantidade de Bolsas', fontsize=12)

#Adicionar os valores acima das barras
for bar in bars:
    yval = bar.get_height()  # Obtém a altura de cada barra
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, str(int(yval)), ha='center', va='bottom', fontsize=12)

#Exibir o gráfico
plt.grid(True, axis='y', linestyle='--', alpha=0.7)  # Adiciona grade horizontal
plt.xticks(rotation=45, ha="right")  # Rotaciona os rótulos do eixo X para melhor visualização
plt.show()
```

```python
#Contar a quantidade de bolsas por área de conhecimento
quantidade_bolsas_area = dados_df['AreaConhecimento'].value_counts()

#Criar o gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(quantidade_bolsas_area.index, quantidade_bolsas_area.values, color='lightgreen')

#Adicionar título e rótulos aos eixos
plt.title('Quantidade de Bolsas por Área de Conhecimento', fontsize=14)
plt.xlabel('Área de Conhecimento', fontsize=12)
plt.ylabel('Quantidade de Bolsas', fontsize=12)

#Adicionar os valores acima das barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.1, str(int(yval)), ha='center', va='bottom', fontsize=12)

#Exibir o gráfico
plt.xticks(rotation=45, ha="right")  # Rotaciona os rótulos do eixo X para melhor visualização
plt.grid(True, axis='y', linestyle='--', alpha=0.7)  # Adiciona grade horizontal
plt.show()
```

```python
#Remover a parte da hora ('00:00') antes de converter para datetime
dados_df['Inicio'] = dados_df['Inicio'].str.split(' ').str[0]

#Converter a coluna 'Inicio' para o formato datetime
dados_df['Inicio'] = pd.to_datetime(dados_df['Inicio'], format='%d/%m/%Y')
```

```python
#Agrupar por ano e mês (usando .dt.to_period('M')) para contar as bolsas por mês
quantidade_bolsas_mes = dados_df.groupby(dados_df['Inicio'].dt.to_period('M')).size()

#Criar o gráfico de linhas
plt.figure(figsize=(10, 6))
plt.plot(quantidade_bolsas_mes.index.astype(str), quantidade_bolsas_mes.values, marker='o', linestyle='-', color='b')

#Adicionar título e rótulos aos eixos
plt.title('Evolução das Bolsas por Mês', fontsize=14)
plt.xlabel('Mês/Ano', fontsize=12)
plt.ylabel('Quantidade de Bolsas', fontsize=12)

#Adicionar os valores sobre cada ponto
for i, valor in enumerate(quantidade_bolsas_mes.values):
    plt.text(i, valor, str(valor), ha='center', va='bottom', fontsize=10, color='black')

#Melhorar a legibilidade dos rótulos do eixo X
plt.xticks(rotation=45, ha='right')

#Exibir o gráfico
plt.grid(True)
plt.tight_layout()  # Ajusta o layout para evitar sobreposição
plt.show()
```
